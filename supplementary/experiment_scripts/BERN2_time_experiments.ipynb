```python
import requests
import time
import json

def query_pmid(pmids, url="http://bern2.korea.ac.kr/pubmed"):
    response = requests.get(url + "/" + ",".join(pmids))
    return response.json(), response.elapsed.total_seconds() * 1000  # Convert seconds to milliseconds

def batch_pmids(pmids, batch_size=100):
    for i in range(0, len(pmids), batch_size):
        yield pmids[i:i + batch_size]

def process_pmids(pmids, batch_size=100):
    all_results = []
    batch_times = []
    batch_index = 0

    for batch in batch_pmids(pmids, batch_size):
        result, elapsed_time = query_pmid(batch)
        all_results.extend(result)
        batch_times.append(elapsed_time)
        print(f"Batch {batch_index + 1} runtime: {elapsed_time:.2f} ms") # Print runtime for batch
        batch_index += 1

    return all_results, batch_times

def load_pmids_from_file(filename):
    with open(filename, 'r') as file:
        pmids = [line.strip() for line in file]
    return pmids

def save_results(results, filename):
    with open(filename, 'w') as f:
        json.dump(results, f, indent=4)

# Time the query and saving process
start_time = time.time()

# Load PMIDs from file
pmid_file = '/content/drive/MyDrive/bern2/pmids.txt'  # The text file with one PMID per row
pmids = load_pmids_from_file(pmid_file)

# Define batch size
batch_size = 5

# Process PMIDs
all_results, batch_times = process_pmids(pmids, batch_size)

# Save results to file
save_results(all_results, '/content/drive/MyDrive/bern2/bern2_annotations.json')

# Calculate total runtime in milliseconds
end_time = time.time()
total_runtime = (end_time - start_time) * 1000

# Print total runtime
print(f"Total runtime: {total_runtime:.2f} ms")
```

# Convert Lund COVID-19 corpus from xml to free text

```python
import xml.etree.ElementTree as ET

# Paths
xml_file = '/content/drive/MyDrive/bern2/Supplemental_file5.xml'  # Replace with the correct path to the uploaded XML file
output_file = '/content/drive/MyDrive/bern2/Lund_COVID19_freetext.txt'  # Output text file path

# Parse the XML file
tree = ET.parse(xml_file)
root = tree.getroot()

# Open the output file for writing
with open(output_file, 'w', encoding='utf-8') as f:
    for document in root.findall('document'):
        title = None
        abstract = None
        for passage in document.findall('passage'):
            infon_type = passage.find('infon[@key="type"]').text
            if infon_type == 'title':
                title = passage.find('text').text
            elif infon_type == 'abstract':
                abstract = passage.find('text').text

        if title:
            f.write(title.rstrip() + '\n')
        if abstract:
            f.write(abstract.rstrip() + '\n')
        f.write('\n')

print("Extraction and merging completed successfully.")
```
