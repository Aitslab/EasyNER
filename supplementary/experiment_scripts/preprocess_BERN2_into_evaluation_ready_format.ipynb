# This script is used for BERN2 preprocessing

The BERN2 annotations available on HunFLAIR2 repository lacks the full text. This small script was used to add raw text also available in the repository to the BERN2 annotations and match the format of all other annotated files.

```python
from bioc import pubtator
```

```python
# infile_annotations_bern = "../annotations/bern/bioid.txt"
# infile_annotations_bern_gs = "../annotations/raw/bioid_text.txt"
infile = "../annotations/bern/tmvar_v3.txt"
infile_gs = "../annotations/raw/tmvar_v3_text.txt"
```

```python
l = []
with open(infile_gs, encoding="utf8") as f:
    docs = pubtator.load(f)
docs
```

```python
str(docs[0])
```

```python
l = []
with open(infile, encoding="utf8") as f:
    lines = f.readlines()
    l2 = []
    for line in lines:
        
        if len(line.strip().split())!=0:
            l2.append(line)
        else:
            
            l.append(l2)
            l2=[]
l
```

```python
print(len(l), len(docs))
```

```python
outfile = "../annotations2/bern_processed/tmvar_v3"
with open(outfile, "w", encoding="utf8") as f:
    for i, d in zip(l,docs):
        f.write(str(d))
        
        for line in i:
            f.write(line)
        f.write("\n")
        
```
